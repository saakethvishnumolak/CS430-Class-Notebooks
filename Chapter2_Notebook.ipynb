## Stratified Sampling

### Let's take `median_income` is very important to predict housing prices

```python
housing['median_income'].hist()
```

#### Most median income are clustered from 1.5 to 6. Sub categorizing the data into 5 categories (labeled 1 to 5): catagory range from 0 to 1.5 (less than 15,000), 1.5 to 3, and so on

```python
housing['income_cat'] = pd.cut(housing['median_income'], bins = [0, 1.5, 3, 4.5, 6, np.inf], labels = [1, 2, 3, 4, 5])
```

```python
housing['income_cat'].hist()
```

```python
housing.info()
```

```python
from sklearn.model_selection import train_test_split
strat_training_test, strat_test_set = train_test_split(housing, test_size = .2, stratify = housing['income_cat'], random_state = 42)
```

```python
strat_training_test.info()
```

### Remove income_cat

```python
for cat in (strat_training_test, strat_test_set):
    cat.drop("income_cat", axis = 1, inplace = True)
```

```python
strat_training_test.info()
```

```python
housing_new = strat_training_test.copy()
```

```python
housing_new.info()
```

## Data Cleaning

### Imputing missing values

```python
housing_new.info()
```

```python
housing_new.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'median')
```

```python
housing_new['total_bedrooms'] = imputer.fit_transform(housing_new[['total_bedrooms']])
```

```python
housing_new.info()
```

### One hot shot encoding using dummy variable

```python
housing_new.columns
```

```python
housing_new = pd.get_dummies(housing_new[['longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income',
       'median_house_value', 'ocean_proximity']], drop_first = True)
```

```python
housing_new.head()
```

```python
housing_new.to_csv('updated_housing.csv', index = False)
```
